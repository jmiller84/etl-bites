Import librarys and load environment variables

```python
import requests
from psycopg2 import connect, sql
import os
from dotenv import load_dotenv #pip install python-dotenv
from os import environ as env
import pandas as pd

# Load environment variables
# Connection string stored in .env file
load_dotenv()
conn_string = os.getenv('conn_string')

if 'conn_string' in env:
    print(env['conn_string'][:35])
```

Using the response module download the data from the API

```python
def get_data_from_api(url):
    response = requests.get(url)
    return response.json()

todos_url = "https://jsonplaceholder.typicode.com/todos"
users_url = "https://jsonplaceholder.typicode.com/users"

todos_data = get_data_from_api(todos_url)
users_data = get_data_from_api(users_url)
```

Join the todos to the users where the todo userId matches the user id

```python
def join_todos_and_users(todos_data, users_data):
    for todo in todos_data:
        for user in users_data:
            if todo['userId'] == user['id']:
                todo['author'] = user['name']
    return todos_data

combined_data = join_todos_and_users(todos_data, users_data)
combined_data
```

Convert joint dictionary to Dataframe

```python


combined_df = pd.DataFrame(combined_data)
combined_df.head()
```

Isolate only the completed todos

```python
completed= combined_df.loc[combined_df['completed'] == True]
completed
```

Count the number of completed todos per user using the value_counts method

```python
completed_per_user = pd.DataFrame(completed.value_counts('author').reset_index('author'))
completed_per_user
```

Create an SQL table to store the new DataFrame

```python
# Create tables in analytical DB
# This could also be done manually via a GUI (e.g. TablePlus) or with a SQL script
def execute_query_postgresql(conn_string, query):
    with connect(conn_string) as conn:
        with conn.cursor() as cur:
            cur.execute(query)
            conn.commit()

create_api_data_table = '''
DROP TABLE IF EXISTS completed_todos CASCADE;
CREATE TABLE completed_todos (
user_name TEXT NOT NULL,
num_posts INTEGER NOT NULL
);
'''

execute_query_postgresql(conn_string, create_api_data_table)
```

Insert dataframe data into the new SQL table

```python
def insert_data_to_postgresql(conn_string, table_name, data):
    with connect(conn_string) as conn:
        with conn.cursor() as cur:
            for row in data.index:
                query = sql.SQL("INSERT INTO {} (user_name, num_posts) VALUES (%s, %s)").format(sql.Identifier(table_name))
                cur.execute(query, (data['author'][row], int(data['count'][row])))
        conn.commit()

table_name = "completed_todos"
insert_data_to_postgresql(conn_string, table_name, completed_per_user)
```
